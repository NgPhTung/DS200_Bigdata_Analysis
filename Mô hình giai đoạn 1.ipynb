```python
import pandas as pd

# Define a function to load data and rename the last column
def load_and_rename(url, nrows, col_index, new_col_name):
    data = pd.read_csv(url, nrows=nrows)
    data = data.iloc[:, [col_index]]
    data.columns = [new_col_name]
    return data

# Load and rename the data from different sources
url_books = '/kaggle/input/amazon-books-reviews/Books_rating.csv'
data_books = load_and_rename(url_books, 400, -1, 'review')

url_grocery = '/kaggle/input/amazon-reviews-dataset/amazon_grocery_Data.csv'
data_grocery = load_and_rename(url_grocery, 300, -3, 'review')

url_jewelry = '/kaggle/input/amazon-reviews-dataset/amazon_jwellery_Data.csv'
data_jewelry = load_and_rename(url_jewelry, 300, -3, 'review')

url_pc = '/kaggle/input/amazon-reviews-dataset/amazon_pc_Data.csv'
data_pc = load_and_rename(url_pc, 300, -3, 'review')

url_train = '/kaggle/input/hierarchical-text-classification/train_40k.csv'
data_train = load_and_rename(url_train, 1300, -4, 'review')

# Concatenate all data into one DataFrame
combined_data = pd.concat([data_books, data_grocery, data_jewelry, data_pc, data_train], ignore_index=True)

# Add a column 'labels' with all values set to 0
combined_data['label'] = 0

# Display the combined DataFrame
print(combined_data)
```

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/IsakZhang/ABSA-QUAD/master/data/rest16/train.txt'
url1 = 'https://raw.githubusercontent.com/IsakZhang/ABSA-QUAD/master/data/rest16/dev.txt'
url2 = 'https://raw.githubusercontent.com/IsakZhang/ABSA-QUAD/master/data/rest16/test.txt'

# Load the dataset with the first row as header
data = pd.read_csv(url, header=0, sep='####', engine='python')
data1 = pd.read_csv(url1, header=0, sep='####', engine='python')
data2 = pd.read_csv(url2, header=0, sep='####', engine='python')

data = data2



# Rename the first column to 'review'
data.rename(columns={data.columns[0]: 'review'}, inplace=True)
#data = data.iloc[:, 0]
# Add a new column 'label' with all values set to 1
data['label'] = 1

selected_data2 = data[['review', 'label']]
selected_data2.isna().sum()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# Load the dataset
# Assuming your dataset is stored in a DataFrame called df with columns 'review' and 'label'
# Preprocessing steps such as cleaning the text may be necessary before proceeding further

# Split the dataset into features (X) and labels (y)
X = df['review'].values
y = df['label'].values

# Tokenize the text data
max_words = 10000  # Maximum number of words to keep in the tokenizer's vocabulary
tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(X)
X_tokenized = tokenizer.texts_to_sequences(X)

# Pad sequences to ensure uniform length
max_sequence_length = 100  # Maximum sequence length (can be adjusted based on dataset)
X_padded = pad_sequences(X_tokenized, maxlen=max_sequence_length, padding='post')

# Split the dataset into training and testing sets (7:3 ratio)
X_train, X_test, y_train, y_test = train_test_split(X_padded, y, test_size=0.3, random_state=42)
```

```python
y_test
```

```python
# Define the LSTM model
embedding_dim = 100  # Dimension of word embeddings
lstm_units = 128  # Number of LSTM units
num_classes = 2  # Number of classes (binary classification)

model = Sequential([
    Embedding(max_words, embedding_dim, input_length=max_sequence_length),
    LSTM(lstm_units),
    Dense(num_classes, activation='softmax')
])

# Compile the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
epochs = 10  # Number of training epochs
batch_size = 32  # Batch size for training
model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_split=0.1)

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print("Test Loss:", loss)
print("Test Accuracy:", accuracy)
```

```python
# Save the model to HDF5 file
model.save('LSTM_rest.h5')
```

```python
def predict_label(sentence, tokenizer, model):
    # Preprocess the input sentence
    max_sequence_length = 100  # Maximum sequence length (should be the same as used during training)
    sentence_tokenized = tokenizer.texts_to_sequences([sentence])
    sentence_padded = pad_sequences(sentence_tokenized, maxlen=max_sequence_length, padding='post')
    
    # Predict the label
    prediction = model.predict(sentence_padded)
    
    # Convert the predicted probabilities to class labels
    predicted_label = 1 if prediction[0][1] > 0.5 else 0
    
    return predicted_label
```

```python
# Example usage:
sentence = "The lobster knuckles ( special of the day ) were ok , but pretty tasteless"
predicted_label = predict_label(sentence, tokenizer, model)
print("Predicted Label:", predicted_label)
```
