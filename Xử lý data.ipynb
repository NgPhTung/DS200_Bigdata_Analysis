- Số bộ 4 trung bình

- Só phần tử khác NULL trong các bộ 4

```python
import pandas as pd
import ast
data_type = 'train'
print("tập dữ liệu", data_type)
data_path = f"/content/drive/MyDrive/Study/Năm 3/DS200 - Big data/project/res16/{data_type}.txt"
inputs = []
outputs = []
with open(data_path, 'r') as f:
  lines = f.readlines()

for line in lines:
  if line == "":
    continue
  try:
    input, output = line.split('####')
  except:
    print(line.split('####'), line)
  output = eval(output)
  # output_str = [', '.join(i) for i in output]
  # output_str = '; '.join(output_str)
  inputs.append(input)
  outputs.append(output)

#1 Số sample
print("Tống số sample", len(inputs))
#2 Số sample chứa nhiều hơn 1 bộ 4
num_multi_4 = 0
for output in outputs:
  if len(output) > 1:
    num_multi_4 += 1
print("Số sample chứa nhiều hơn 1 bộ 4", num_multi_4)
#3 Tổng số bộ 4 trong outputs
single_quad = []
for output in outputs:
    single_quad.extend(output)
print("Tổng số bộ 4", len(single_quad))
#4 Số bộ 4 tối đa, số bộ 4 tối thiểu, bộ 4 trung bình trong outputs
max_quad = max(len(output) for output in outputs)
min_quad = min(len(output) for output in outputs)
avg_quad = sum(len(output) for output in outputs) / len(outputs)
print("Số bộ 4 tối đa", max_quad)
print("Số bộ 4 tối thiểu", min_quad)
print("Số bộ 4 trung bình", avg_quad)
# Độ dài tối đa của câu, Đối dài tối thiểu cảu câu, Độ dài trung bình
max_length = max([len(i.split(" ")) for i in inputs])
min_length = min([len(i.split(" ")) for i in inputs])
avg_length = sum([len(i.split(" ")) for i in inputs]) / len(inputs)
print("Độ dài tối đa của câu", max_length)
print("Độ dài tối thiểu cảu câu", min_length)
print("Độ dài trung bình", avg_length)

#Số phần tử khác NULL trong các bộ 4
for ele in [0, 1, 2, 3]:
  count_null = 0
  unique_values = set()
  for output in single_quad:
    if len(output) > 0:
      if output[ele] == "NULL":
        count_null += 1
      else:
        unique_values.add(output[ele])
    else:
      print("Câu không có quad")
  print(f"Phần tử {ele} NULL: {count_null}")
  print(f"Phần tử {ele}   unique_value: {len(unique_values)}")
  if len(unique_values) < 15:
    print(f"Phần tử {ele} unique_value: {unique_values}")

# Vẽ biểu đồ Phân khối sentiment
sentiment = [output[2] for output in single_quad]
from collections import Counter
count_sentiment = Counter(sentiment)
print(count_sentiment)
import matplotlib.pyplot as plt
plt.bar(count_sentiment.keys(), count_sentiment.values())
plt.show()
```

```python
import pandas as pd
test_result_df = pd.read_csv("/content/drive/MyDrive/project/res16/res16_test_result.csv")
pred_pt = []
gold_pt = []
for pred_sen in test_result_df['predict']:
  quad_list = pred_sen.split('; ')
  quad_sen_list = [i.split(', ') for i in quad_list]
  pred_pt.append(quad_sen_list)

for pred_sen in test_result_df['y_output']:
  quad_list = pred_sen.split('; ')
  quad_sen_list = [i.split(', ') for i in quad_list]
  gold_pt.append(quad_sen_list)
print("Gold mẫu", gold_pt[0:5])
print("Pred mẫu", pred_pt[0:5])
#   pred_pt.append(ast.literal_eval(test_result_df.iloc[i]['pred_quad']))
# test_result_df.head()
```

```python
def compute_f1_scores(pred_pt, gold_pt):
    """
    Function to compute F1 scores with pred and gold quads
    The input needs to be already processed
    """
    # number of true postive, gold standard, predictions
    n_tp, n_gold, n_pred = 0, 0, 0

    for i in range(len(pred_pt)):
        n_gold += len(gold_pt[i])
        n_pred += len(pred_pt[i])

        for t in pred_pt[i]:
            if t in gold_pt[i]:
                n_tp += 1

    print(f"number of gold spans: {n_gold}, predicted spans: {n_pred}, hit: {n_tp}")
    precision = float(n_tp) / float(n_pred) if n_pred != 0 else 0
    recall = float(n_tp) / float(n_gold) if n_gold != 0 else 0
    f1 = 2 * precision * recall / (precision + recall) if precision != 0 or recall != 0 else 0
    scores = {'precision': precision, 'recall': recall, 'f1': f1}

    return scores

scores = compute_f1_scores(pred_pt, gold_pt)
print(scores)
```
